## List Comphrehesions


```python
array_number = []
for x in range(1, 100 + 1):
    array_number.append(x**2)
print(array_number)
```

```python
array_number_01 = []
for x in range(1, 100 + 1):
    if x % 3 != 0:
        array_number_01.append(x**2)
print(array_number_01)
```

```python
array_number_02 = [i**2 for i in range(1, 100 + 1) if i % 3 != 0]
print(array_number_02)
```

## Dictionary comprehensions


```python
dict_number_01 = {}

for i in range(1, 100 + 1):
    dict_number_01[i] = i**3

print(dict_number_01)
```

```python
dict_number_02 = {}

for i in range(1, 100 + 1):
    if i % 3 != 0:
        dict_number_02[i] = i**3

print(dict_number_02)
```

```python
dict_number_03 = {i: i**3 for i in range(1, 100 + 1) if i % 3 != 0}
print(dict_number_03)
```

```python
dict_number_04 = {i: i**0.5 for i in range(1, 10 + 1)}
print(dict_number_04)
```

## Funciones anónimas: lambda


```python
palidrome = lambda string: string == string[::-1]
print(palidrome("Luis"))
print(palidrome("ana"))
```

High order functions: filter, map y reduce


```python
# filter
array_a = [1, 2, 3, 4, 5]
rpta = list(filter(lambda x: x < 4, array_a))
print("Filter:")
print(rpta)
# map
array_a = [1, 2, 3, 4, 5]
rpta = list(map(lambda x: x**2, array_a))
print("map:")
print(rpta)
# Reduce
from functools import reduce
array_a = [1, 2, 3, 4, 5]
rpta = reduce(lambda x, y: x * y, array_a)
print(rpta)
```
